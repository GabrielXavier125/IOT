```python
# -*- coding: utf-8 -*-
# Cdigo adaptado para VS Code e compatvel com a base Sample Superstore

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurao visual
sns.set_theme(style="darkgrid")

# Carregar base de dados
df = pd.read_csv('SampleSuperstore.csv')

# Visualizar primeiras linhas
print(df.head())

# Verificando tipos de dados e nulos
print(df.info())

# ---------------------------------------------------------
# 1) DISTRIBUIO DE VENDAS
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.histplot(df["Sales"], bins=30, kde=True, color='skyblue')
plt.title("Distribuio das Vendas (Sales)")
plt.xlabel("Valor das Vendas")
plt.ylabel("Contagem")
plt.show()

# ---------------------------------------------------------
# 2) DISTRIBUIO DE LUCRO (Profit)
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.histplot(df["Profit"], bins=30, kde=True, color='green')
plt.title("Distribuio do Lucro (Profit)")
plt.xlabel("Lucro")
plt.ylabel("Contagem")
plt.show()

# ---------------------------------------------------------
# 3) BOXPLOT - LUCRO POR CATEGORIA
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Category", y="Profit", palette="Set2")
plt.title("Lucro por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Lucro")
plt.show()

# ---------------------------------------------------------
# 4) KDE - Densidade das Vendas
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x="Sales", fill=True, color="orange")
plt.title("Densidade das Vendas (Sales)")
plt.xlabel("Valor das Vendas")
plt.show()

# ---------------------------------------------------------
# 5) HISTOGRAMA DE VENDAS POR REGIO
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Sales", hue="Region", multiple="stack", bins=25, palette="pastel")
plt.title("Distribuio de Vendas por Regio")
plt.xlabel("Vendas")
plt.show()

# ---------------------------------------------------------
# 6) BOXPLOT - VENDAS POR SEGMENTO
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Segment", y="Sales", palette="Set3")
plt.title("Vendas por Segmento")
plt.xlabel("Segmento")
plt.ylabel("Vendas")
plt.show()

# ---------------------------------------------------------
# 7) VIOLINPLOT - LUCRO POR SUB-CATEGORIA
# ---------------------------------------------------------
plt.figure(figsize=(14, 7))
sns.violinplot(data=df, x="Sub-Category", y="Profit", palette="muted")
plt.xticks(rotation=45)
plt.title("Distribuio do Lucro por Sub-Categoria")
plt.show()
```
